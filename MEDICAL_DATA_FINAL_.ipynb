```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df_train = pd.read_csv('train_df.csv')
```

```python
df_test = pd.read_csv('test_df.csv')
```

```python
display(df_train.head())
display(df_test.head())
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
# checking any missing values
df_train.isnull().sum()
```

```python
# checking any missing values
df_test.isnull().sum()
```

```python
categorical_cols = df_train.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    df_train[col] = df_train[col].astype('category').cat.codes
    df_test[col] = df_test[col].astype('category').cat.codes
```

```python
# Exploratory Data Analysis
plt.figure(figsize=(10, 5))
sns.histplot(df_train['age'], bins=25, kde=True)
plt.title('Age Distribution of Patients (Training Data)')
plt.show()
```

```python
# Readmission Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='readmitted', data=df_train)
plt.title('Readmission Distribution (Training Data)')
plt.show()
```

```python
# Readmission Count Plot
if 'readmitted' in df_train.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='readmitted', data=df_train)
    plt.title('Readmission Distribution (Training Data)')
    plt.show()
else:
    print("Column 'Readmitted' not found in dataset.")
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df_train.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix (Training Data)')
plt.show()
```
